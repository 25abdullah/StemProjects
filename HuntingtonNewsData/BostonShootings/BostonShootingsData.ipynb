```python
import pandas as pd
data_frame = pd.read_csv('Boston_shooting_data_set.csv')
data_frame['shooting_date'] = pd.to_datetime(data_frame['shooting_date'])
filter_2015_fatal = ((data_frame['shooting_date'].dt.year == 2015) &
 (data_frame['shooting_type_v2'] == 'Fatal'))
length_fatal_2015 = len(data_frame[filter_2015_fatal])
filter_2015_non_fatal = ((data_frame['shooting_date'].dt.year == 2015) &
 (data_frame['shooting_type_v2'] == 'Non-Fatal'))
length_non_fatal_2015 = len(data_frame[filter_2015_non_fatal])
print(length_fatal_2015) #outputs 33
print(length_non_fatal_2015) #outputs 212
total_shootings_2015 = (length_fatal_2015 + length_non_fatal_2015)
print(total_shootings_2015) #outputs 245

data_frame.head()
```

```python
#percent change formula: (new - old )/ old *100
#I used 2024's data because 2025 is not over
data_frame['shooting_date'] = pd.to_datetime(data_frame['shooting_date'])
filter_2024_fatal = ((data_frame['shooting_date'].dt.year == 2024) &
 (data_frame['shooting_type_v2'] == 'Fatal'))
length_fatal_2024 = len(data_frame[filter_2024_fatal])
filter_2024_non_fatal = ((data_frame['shooting_date'].dt.year == 2024) &
 (data_frame['shooting_type_v2'] == 'Non-Fatal'))
length_non_fatal_2024 = len(data_frame[filter_2024_non_fatal])
#yay! Non-fatal = 109, fatal = 20 = 129 total shootings
print(length_fatal_2024) #outputs 20
print(length_non_fatal_2024) #outputs 109
total_shootings_2024 = (length_fatal_2024 + length_non_fatal_2024)
print(total_shootings_2024) #outputs 129
change_percent = ((total_shootings_2024 - total_shootings_2015)
/total_shootings_2015 *100)
print(change_percent) #we see a reduction of 47% since 2016 in shootings
print("we see a reduction of " + str(abs(change_percent)) +  "% in total shootings from 2015 - 2024.")


```

```python
#counting number of fatal shootings and number of non-fatal shootings over all time
filter_fatal = (data_frame['shooting_type_v2'] == 'Fatal')
data_frame_fatal_length = len(data_frame[filter_fatal])
print(str(data_frame_fatal_length) + " is the number of fatal shootings from 2015 - 2024.")
filter_non_fatal = (data_frame['shooting_type_v2'] == 'Non-Fatal')
data_frame_non_fatal_length = len(data_frame[filter_non_fatal])
print (str(data_frame_non_fatal_length) + " is the number of non-fatal shootings from 2015 - 2024.")

#total shootings
total_shootings = (data_frame_fatal_length + data_frame_non_fatal_length) #2141
print("Between 2015 and 2024, there have been " + str(total_shootings) +
      "shootings.")


#percent of fatal shootings
fatal_percent = (335/total_shootings)*100
print("Percentage of fatal shootings from 2015 - 2024 is " + str(fatal_percent))
non_fatal_percent = (1786/total_shootings)*100
print( "Percentage of shootings that are non-fatal all time from 2015-2024 " +
str(non_fatal_percent))

```
